# Prime Factorization

```python
def prime(m):
    list=[]
    b=m+1
    for i in range(1,b):
        if m%i==0:
            list.append(i)
        else: continue
    if len(list)>2:
        return False
    else:
        return True    
        
        
     
    
    

n=int(input("Type a number: "))
prod=1
list_divisors=[]
b=n+1
for i in range(1,b):
    if n%i==0:
        list_divisors.append(i)
    else: continue
list_prime_factors=[]
for i in list_divisors:
    if prime(i)==True:
        prod=prod*i
        list_prime_factors.append(i)
    else: continue
print(list_prime_factors,"is the list of prime factors of",n)

list_factors=[]
for j in list_prime_factors:
    if j!=1:
        for k in range(1,50):
            if n%(j**k)==0:
                list_factors.append(j)
            else: continue
    else: continue
list1=[]
for s in list_factors:
    list1.append(str(s))
t="x".join(list1)
print("Prime Factorization of",n,"is",t,"=",n)
   
        
        
        
```
